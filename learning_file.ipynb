```python
# Encode categorical variables
label_encoders = {}
for column in ['Cinsiyet', 'Meslek', 'Mutluyken', 'Üzgünken', 'Öfkeliyken']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
features = df[['Yaş', 'Cinsiyet', 'Meslek']]
labels_happy = df['Mutluyken']
labels_sad = df['Üzgünken']
labels_angry = df['Öfkeliyken']
```

```python
# Splitting the data
X_train_happy, X_test_happy, y_train_happy, y_test_happy = train_test_split(features, labels_happy, test_size=0.2, random_state=42)
X_train_sad, X_test_sad, y_train_sad, y_test_sad = train_test_split(features, labels_sad, test_size=0.2, random_state=42)
X_train_angry, X_test_angry, y_train_angry, y_test_angry = train_test_split(features, labels_angry, test_size=0.2, random_state=42)
```

```python
# Training and evaluating the model for each emotional state
def train_and_evaluate(X_train, X_test, y_train, y_test, label_encoder, state):
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    target_names = [str(label) for label in label_encoder.classes_]
    print(f"{state} State Classification Report (with target names):")
    print(classification_report(y_test, y_pred, target_names=target_names, labels=label_encoder.transform(label_encoder.classes_), zero_division=0))
    
    return model
modelHappy = train_and_evaluate(X_train_happy, X_test_happy, y_train_happy, y_test_happy, label_encoders['Mutluyken'], 'Happy')
modelSad = train_and_evaluate(X_train_sad, X_test_sad, y_train_sad, y_test_sad, label_encoders['Üzgünken'], 'Sad')
modelAngry = train_and_evaluate(X_train_angry, X_test_angry, y_train_angry, y_test_angry, label_encoders['Öfkeliyken'], 'Angry')
```

```python
import joblib

joblib.dump(modelHappy, 'models/rfHappyModel.pkl')
joblib.dump(modelSad, 'models/rfSadModel.pkl')
joblib.dump(modelAngry, 'models/rfAngryModel.pkl')

# Save the label encoders
joblib.dump(label_encoders['Cinsiyet'], 'models/propsModels/le_gender.pkl')
joblib.dump(label_encoders['Meslek'], 'models/propsModels/le_job.pkl')
joblib.dump(label_encoders['Mutluyken'], 'models/propsModels/le_happy.pkl')
joblib.dump(label_encoders['Üzgünken'], 'models/propsModels/le_sad.pkl')
joblib.dump(label_encoders['Öfkeliyken'], 'models/propsModels/le_angry.pkl')
```
